### 1.1 Preprocessing Steps
1. Load data_identification.csv and emotion.csv files
2. Combine Dataset and Emotion Labels
- Joins the split info (train/test) with emotion labels using the id as the key.- 
- Result: A DataFrame where each row has:
    * id: Post identifier (e.g., 0x61fc95)
    * split: Either "train" or "test"
    * emotion: The emotion label (only for training posts; NaN for test posts)
3. Load and Parse Raw JSON Posts
4. Convert to DataFrame and Standardize ID Column
- Converts the list of posts into a pandas DataFrame.
- Renames 'post_id' → 'id' to match the column name used in other files (data_identification.csv, emotion.csv), enabling clean merging.
5. Merge Posts with Labels and Splits
- Combines text content and metadata (like text and hashtags) with the split and emotion info.
- Now each row contains:
    * id
    * split
    * emotion (only for training)
    * text: Raw social media post
    * hashtags: List of hashtags (if any)
6. Drop Unnecessary Column
- Hashtags are not used in this. Removing them simplifies the data.
7. Clean Text Data
- Remove URLs: Social media posts often contain links (http://...), which add noise and little emotional signal.
- Normalize whitespace: Replace multiple spaces/tabs/newlines with a single space.
- Handle non-string inputs: Some posts may be NaN or non-text; they are converted to empty strings.
- Output: A new column 'clean_text' with normalized, URL-free text ready for modeling.
8. Split into Train and Test Sets


### 1.2 Feature Engineering Steps

2. Split Training Data into Train and Validation Sets
- Create a validation set (20% of training data) to evaluate model performance during training, without touching the test set.

### 2.1 Mention Different Things You Tried
At the beginning of the project, we encountered several challenges in correctly loading and parsing the raw data. The primary data source, final_posts.json, had a deeply nested structure that was not immediately obvious. Our initial attempts to load the file using standard pd.read_json() or a simple json.load() produced a DataFrame with only one column ('root') containing lists or dictionaries, which made it impossible to access the actual post IDs and text content.
This failed to extract usable fields. Only after inspecting the raw structure manually did we realize that each post was stored under the path post['root']['_source']['post'].

Once data loading was resolved, we initially planned to use traditional text representation methods such as TF-IDF with n-grams and Logistic Regression or SVM—standard baselines for text classification. While these approaches are efficient and interpretable, preliminary experiments showed limited performance, likely because social media text is short, noisy, and rich in emotional nuance that word-based models struggle to capture.

To address this, we shifted toward semantic embeddings using Google’s text-embedding-004 model via the Google Generative AI API. 

I  also experimented with:
- Text cleaning strategies (removing URLs, normalizing user mentions and hashtags),
- Stratified train/validation splits to preserve class distribution,
- Caching embeddings to avoid repeated API calls during development.

Ultimately, the embedding-based pipeline proved both robust and effective, forming the basis of our final submission.

### 2.2 Mention Insights You Gained
Through this process, we gained several key insights:
1. Data structure matters: Even with clean code, misunderstanding the JSON schema can lead to silent failures (e.g., empty or misaligned features). Always inspect raw data before modeling.
2. Semantic embeddings outperform traditional methods for emotion detection in informal text. While TF-IDF works well on formal, topic-driven documents, it lacks the contextual awareness needed to distinguish subtle emotional cues in short social media posts.

```python
import re

def clean_text_col(text):
    if not isinstance(text, str):
        return ""
    
    text = re.sub(r'http\S+|www\S+|https\S+', '', text, flags=re.MULTILINE)
    text = re.sub(r'\s+', ' ', text).strip()
    return text

df_full['clean_text'] = df_full['text'].apply(clean_text_col)
```

```python
import os
import time
import numpy as np
import pandas as pd
from tqdm import tqdm
from dotenv import load_dotenv
from google import genai
from sklearn.preprocessing import LabelEncoder

env_path = "./config/.env"
load_dotenv(dotenv_path=env_path)
api_key = os.getenv("GOOGLE_API_KEY")
client = genai.Client(api_key=api_key)

def get_embeddings_smart(texts, client, filename, batch_size=50):
    print(f"Generating embeddings via API for {len(texts)} texts...")
    embeddings = []
    
    for i in tqdm(range(0, len(texts), batch_size), desc="Embedding Batches"):
        batch_texts = texts[i : i + batch_size]
        
        for attempt in range(3):
            try:
                result = client.models.embed_content(
                    model="models/text-embedding-004",
                    contents=batch_texts
                )
                
                if hasattr(result, 'embeddings'):
                    embedding_list = result.embeddings
                elif hasattr(result, 'embedding'):
                    embedding_list = result.embedding
                else:
                    raise AttributeError("Response has neither 'embedding' nor 'embeddings'")
                
                batch_embeddings = []
                for emb_obj in embedding_list:
                    batch_embeddings.append(list(emb_obj.values))
                embeddings.extend(batch_embeddings)

                time.sleep(0.5)
                break
                
            except Exception as e:
                print(f"Batch {i} error: {e}. Retrying...")
                time.sleep(1.0)
        else:
            print(f"Failed batch {i}. Filling with zeros.")
            embeddings.extend([[0.0] * 768 for _ in range(len(batch_texts))])
    
    embeddings_array = np.array(embeddings, dtype=np.float32)
    
    np.save(filename, embeddings_array)
    return embeddings_array 

le = LabelEncoder()
y_full_encoded = le.fit_transform(df_train['emotion'])

X_train_text, X_val_text, y_train_enc, y_val_enc = train_test_split(
    df_train['clean_text'].tolist(), 
    y_full_encoded,
    test_size=0.2, 
    random_state=42, 
    stratify=y_full_encoded
)

X_test_text = df_test['clean_text'].tolist()

print("\n>>> Phase 1: Embeddings")
X_train_emb = get_embeddings_smart(X_train_text, client, "X_train_emb.npy")
X_val_emb   = get_embeddings_smart(X_val_text,   client, "X_val_emb.npy")
X_test_emb  = get_embeddings_smart(X_test_text,  client, "X_test_emb.npy")
```

## 3. Model Implementation Steps

```python
### Add the code related to the model implementation steps in cells inside this section
```

```python
import os
import numpy as np
import pandas as pd
from sklearn.svm import LinearSVC
from sklearn.model_selection import train_test_split
from sklearn.metrics import f1_score
from tqdm import tqdm
import time

print(">>> Loading cached embeddings...")
X_train_emb = np.load("X_train_emb.npy")
X_val_emb = np.load("X_val_emb.npy")
X_test_emb = np.load("X_test_emb.npy")

le = LabelEncoder()
y_full_encoded = le.fit_transform(df_train['emotion'])
_, _, y_train_enc, y_val_enc = train_test_split(
    y_full_encoded, y_full_encoded, test_size=0.2, random_state=42, stratify=y_full_encoded
)

print("\nTraining LinearSVC...")

base_svc = LinearSVC(class_weight='balanced', dual=False, random_state=42, max_iter=5000)

start = time.time()
base_svc.fit(X_train_emb, y_train_enc)  # Train on 80% for validation
print(f"Trained in {time.time() - start:.1f} seconds.")

y_val_pred = base_svc.predict(X_val_emb)
val_f1 = f1_score(y_val_enc, y_val_pred, average='macro')
print(f"Macro F1 (no calibration): {val_f1:.4f}")

X_full = np.concatenate([X_train_emb, X_val_emb])
y_full = np.concatenate([y_train_enc, y_val_enc])
base_svc.fit(X_full, y_full)

test_preds = base_svc.predict(X_test_emb)
final_preds = le.inverse_transform(test_preds)

submission = pd.DataFrame({
    'id': df_test['id'],
    'emotion': final_preds
})
submission.to_csv('submission_linearsvc.csv', index=False)
print("Saved 'submission_linearsvc.csv'")
```
